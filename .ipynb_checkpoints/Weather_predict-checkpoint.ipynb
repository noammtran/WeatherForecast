```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay,precision_score, recall_score, f1_score, accuracy_score, confusion_matrix, mean_squared_error, classification_report
from sklearn.compose import ColumnTransformer
import joblib
```

```python
#read the dataset 
df = pd.read_csv("retrievebulkdataset.csv")

#remove the unnecessary column of data from the dataframe 
df = df.drop(['name','datetime','sunrise','sunset','moonphase', 'description' ,'icon','stations', 'snowdepth' ,'snow','severerisk'], axis=1)

#Check for null data
print("NaN values per column:")
print(df.isna().sum())

#check for preciptype data
print(df['preciptype'].value_counts(normalize=True))

#preciptype column have 2 value is rain and null, is not good for the training so we change the null to no_rain
df['preciptype'] = df['preciptype'].fillna('no_rain')
```

```python
#print out all the possible value of the "conditions" column, which is the output of
print(df['conditions'].value_counts(normalize=True))
```

```python
#our dataframe has some columns with categorical data, so we need to encode that into a numeric type

#in this section, we choose to use Label encode 
le = LabelEncoder()


X_original = df.copy()
X_original['conditions']= le.fit_transform(df['conditions'])
X_original['preciptype']= le.fit_transform(df['preciptype'])

#made a heatmap to see the correlation between each data
#In the heatmap, the closer the value is to 0, the weaker the correlation.
plt.figure(figsize=(12, 8))
sns.heatmap(X_original.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
data=df.drop(['winddir'], axis = 1) # we drop the column winddir since its correlation with conditions is too weak

#Make input X and output y for the training
X = data.drop('conditions', axis=1)
# use Label encoding to output(conditions) since its value is categorical type
y = le.fit_transform(data['conditions'])

#Divide the input into 2 part, numeric and categorical, for later use
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

#made a processor to standadize the numeric data and One-hot encoding the categorical data
preprocessor = ColumnTransformer(
    transformers=[
        # We standardize the data to help the model process better and faster, if we don't use it, the model still works but less efficiently
        ('num', StandardScaler(), numeric_features), 
        # In this case, if we use Label encoding, the model can think some value can be smaller than the other (Label encode is mark the number from 0->n)
        # For example rain = 0; no_rain = 1, with this, the model may think rain is bigger than no_rain then make some mistakes on training
        # So in this case, we use One-hot encoding to encode it into vector
        ('encode', OneHotEncoder(sparse_output=False), categorical_features) 
    ]
)
X = preprocessor.fit_transform(X)
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

#split the data into train data, validation data, test data
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=18)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.2, random_state=18)


best_model = None
best_score = 0

#checking to see which model is the best with different regularization strengths
for c in [0.001, 0.01, 0.1, 1, 10,100]: # c=1/lambda
    model = OneVsRestClassifier(LogisticRegression(solver='newton-cg',C=c, max_iter=1000000))
    model.fit(X_train, y_train)

    val_score = accuracy_score(y_val, model.predict(X_val))
    print(f"C={c} => Validation Accuracy: {val_score:.3f}")

    if val_score > best_score:
        best_score = val_score
        best_model = model

test_score = accuracy_score(y_test, best_model.predict(X_test))
print(f"\nFinal Test Accuracy with best model: {test_score:.3f}")
```

```python
# Get test predictions
y_test_pred = best_model.predict(X_test)
labels = np.unique(np.concatenate((y_test, y_test_pred)))

# Compute confusion matrix
cm = confusion_matrix(y_test, y_test_pred, labels=labels)

# Display it
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_[labels])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - Test Set")
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Accuracy
accuracy = accuracy_score(y_test, y_test_pred)

# 2. Precision, Recall, F1-Score (macro average: treats all classes equally)
precision = precision_score(y_test, y_test_pred, average='macro', zero_division=0)
recall = recall_score(y_test, y_test_pred, average='macro', zero_division=0)
f1 = f1_score(y_test, y_test_pred, average='macro', zero_division=0)

# 4. Specificity (needs per-class confusion matrix)
FP = cm.sum(axis=0) - np.diag(cm)
TN = cm.sum() - (cm.sum(axis=1) + cm.sum(axis=0) - np.diag(cm))
specificity = np.mean(TN / (TN + FP + 1e-10))

# Display all
print(f"Accuracy     : {accuracy:.3f}")
print(f"Precision    : {precision:.3f}")
print(f"Recall       : {recall:.3f}")
print(f"F1-Score     : {f1:.3f}")
print(f"Specificity  : {specificity:.3f}")
```

```python
#Save the model
joblib.dump(best_model, 'model.pkl')
joblib.dump(preprocessor, 'preprocessor.pkl')
joblib.dump(le, 'label_encoder.pkl')
```
